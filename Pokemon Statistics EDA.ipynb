There are 800 pokemons in the given dataset
RayquazaMega Rayquaza,MewtwoMega Mewtwo Y and MewtwoMega Mewtwo X are the most powerfull pokemons with a common overall stat of 780 

# Top 10 pokemons

```python
#Top 10 pokemons
df[["Name","Total"]].sort_values(by= "Total",ascending= False).head(10)
```

# Sunkern is the least powerfull pokemon with lowest total stat of 180 followed by Azurill

```python
#Least 10 powerfull pokemons
df[["Name","Total"]].sort_values(by= "Total",ascending= False).tail(10)
```

```python
Water type pokemons are the strongest type of pokemons and  Flying type pokemons are least strongest
```

```python
x= df.groupby("Type 1").Total.sum().sort_values(ascending= False)
fig= sns.barplot(x.values,x.index)
fig.set_xlabel("Overall total")
fig.set_ylabel("Type")
```

# fastest types of pokemon

The water type pokemons are also the fastest and most defensive types of pokemons

```python
#fastest types of pokemon
x= df.groupby("Type 1").Speed.sum().sort_values(ascending= False)

fig= sns.barplot(x.values,x.index)
fig.set_xlabel("Speed")
fig.set_ylabel("Type")
```

```python
#Most defensive types of pokemon
y= df.groupby("Type 1").Defense.sum().sort_values(ascending= False)
fig= sns.barplot(y.values,y.index)
fig.set_xlabel("Defense")
fig.set_ylabel("Type")
```

# DeoxysSpeed Forme is the fastest pokemon

```python
#Top 10 fastest pokemons
df[["Name","Speed"]].sort_values(by= "Speed",ascending= False).head(10)
```

# MewtwoMega Mewtwo X has best attack follwed by HeracrossMega Heracross 

```python
#pokemons with best attacks
df[["Name","Attack"]].sort_values(by= "Attack",ascending= False).head(10)
```

# Attack and defense correlation

```python
#Attack and defense correlation
plt.figure(figsize=(10,10))
sns.lmplot(data= df,x= "Attack",y= "Defense")
```

# Number of pokemons in each type 1

```python
x= df["Type 1"].value_counts()
sns.barplot(x.values,x.index)
```

# Number of pokemon per generation

```python
x= df.Generation.value_counts()
ax= sns.barplot(x.index,x.values)
ax.set_xlabel("Generation")
ax.set_ylabel("No of pokemons")
```

# Violin plot distribution of type 1 pokemons based on attack

```python
plt.figure(figsize=(20,10))
ax= sns.violinplot(x= "Type 1",y= "Attack",data= df)
```

# Violin plot distribution of type 1 pokemons based on speed

```python
plt.figure(figsize=(20,10))
ax= sns.violinplot(x= "Type 1",y= "Speed",data= df)
```

# Legendary pokemon distribution


```python
x= df.Legendary.value_counts()
labels=["Non legendary","Legendary"]
plt.pie(x,labels= labels,autopct= "%1.1f%%")
plt.show()
```

# Type 2 pokemon distribution pie chart

```python
x= df["Type 2"].value_counts()
labels= list(x.index)
plt.pie(x,labels= labels)
plt.plot
```

# Boxplot for overall total for generations of pokemon

```python
plt.figure(figsize=(10,10))
sns.boxplot(x= df.Generation,y= df.Total)
```
